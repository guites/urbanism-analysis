```python
import json
import pandas as pd
import requests
import urllib
```

```python
# Carregar csv com população por bairro
bairros = pd.read_csv("floripa-bairros.csv")
```

```python
bairros.head()
```

```python
# URL base da API OPS
API_URL = "https://nominatim.openstreetmap.org/search.php"

# save formatted geojson data
geojson = {
    "type": "FeatureCollection",
    "features": []
}

# Para cada bairro, pegas sua área limite em formato geojson
for bairro in bairros['Bairro']:
    payload = {
        "q": f"{bairro}, Florianópolis, Santa Catarina, Brazil",
        "type": "administrative",
        "polygon_geojson": 1,
        "format": "jsonv2"
    }
    r = requests.get(API_URL, params=payload)
    if r.status_code != 200:
        continue
    received = r.json()
    if len(received) == 0:
        continue
    top_suggestion = received[0]
    formatted_geojson = {
        "type": "Feature",
        "properties": {
            "id": str(top_suggestion['place_id']),
            "name": bairro,
            "description": top_suggestion['display_name'],
            "geometry": top_suggestion['geojson']
        }
    }
    geojson["features"].append(formatted_geojson)

with open("bairros-geojson.json", "w") as f:
    json.dump(geojson, f)
```
